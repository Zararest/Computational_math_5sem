### Постановка задачи
Необходимо решить уравенние Ван-дер-Поля с параметрами, соответствующими жесткой задаче:

$$ 
\frac{d y_1}{dt} = y_1 \\
\frac{d y_2}{dt} = -e (y_2 (y_1^2 - 1) + y_1) 
$$

Параметры системы:

$$
e = 10^3 \\
e = 10^6
$$

Для решения такой системы был выбран метод Розенброка 3 порядка

### Условие жесткости системы y'=f(y)
1. Собственные значения матрицы Якоби f(y) имеют отрицательные действиетельные части
2. число жесткости $ S = \frac{max(Re \lambda_k)}{min(Re \lambda_m)} $ велико

### Метод Розенброка 3 порядка
Данный метод предназначен для решения системы автономных дифференциальных уравнений:

$$
y' = f(y)
$$

$$
y_{n + 1} = y_n + p_1 k_1 + p_2 k_2 + p_3 k_3 \\
\begin{cases}
(E + a h f'(y_n)) k_1 = h f(y_n)                                  \\
(E + a h f'(y_n)) k_2 = h f(y_n + \beta_{21} k_1)                 \\
(E + a h f'(y_n)) k_3 = h f(y_n + \beta_{31} k_1 + \beta_{32} k_2)\\
\end{cases}  
$$

Для данной системы:

$$
\begin{equation}
f = 
\begin{pmatrix}
  y_2                      \\
  - e (y_2(y_1^2 - 1) + y_1)
\end{pmatrix}
\end{equation}
$$

$$
\begin{equation}
f'(y_n) = 
\begin{pmatrix}
  0                        & 1     \\
  -2 e {y_1}_n {y_2}_n - e & - e({y_1^2}_n - 1)
\end{pmatrix}
\end{equation}
$$

Чтобы решить данную систему, необходимо на каждом шаге решать СЛАУ, поэтому для вычисления $k_i$ будет использоваться метод верхних релаксаций. Матрица A и столбец f будут вычисляться на основе численной производной по формуле:

$ {4 \over 3} {{f(x + h) - f(x - h)} \over {2h}} - {1 \over 3} {{f(x + 2h) - f(x - 2h)} \over {4h}}$
