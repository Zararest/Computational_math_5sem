# Экстраполяция полиномом Ньютона 
По заданной функции был построен полином Ньютона с коэффицикетами:

$$
[ 9.22284960e+07,  1.37930410e+06,  1.69402300e+04, -1.93453133e+03, \\
  1.25106725e+02, -4.15059800e+00,  8.05654167e-02, -7.56485278e-04, \\
 -7.56467535e-06,  5.09932804e-07]
$$

Для вычисления коэффициентов полинома был использован алгоритм с записью промежуточных значений в память. Обход таблицы производился сверху вниз, слева направо.

Экстраполированное значение населения США в 2010 по 10 значениям:

$$
N_{10} = 810 000 000
$$

График:

![Картинку украли цыгане](img/Newton_USA.png)


Данное значение превосходит реальное значение бролее чем в 2 раза. Это связано с тем, что при построении полинома его экстремумы лежат в области между реальным изначениями. Таким образом интерполированные значения достаточно стабильны. При стремлении к бесконечности значения растут пропрционально старшей степени, поэтому в области где нет экстремумов(случай экстраполяции) наблюдается сильный рост модуля функции. 
Таким образом для лучшей экстраполяции необходимо брать меньшее количество точек, поскольку полином будет расти медленне.

Значение населения США в 2010 году по 3 последним точкам:

$$
N_{3} = 318 000 000
$$

График:

![Картинку украли цыгане](img/Newton_USA_3.png)

Данное значение хорошо сходится с реальными данными, поскольку реальная зависимость близка к линейном, поэтому понижение степени полинома дает более точное значение.
