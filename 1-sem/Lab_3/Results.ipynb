### Решение нелинейного уравнения

Уравнение:

$$ 2 \lg{x} - x / 2 + 1 = 0 $$

Построим графики уравнений

$$  y = 2 \lg{x} \\
    y = x / 2 - 1
$$

и найдем примерную точку пересечения.

График:

![Картинку украли цыгане](Approx.png)

Локализация:

$$  x_1 \in [0.2, 1] \\
    x_2 \in [4.5, 5]
$$

В качетве функции $\varphi(x)$ выберем:

$$ \varphi_1(x) = 10^{x / 4 - 1/ 2} $$

ее производная меньше 1 на первой области локализации.
Для вторйо области локализации подходит функция:

$$ \varphi_2(x) = 4 \lg{x} + 2 $$

Корни уравнения при $\varepsilon = 0.001$:

$$  x_1 = 0.3976 \\
    x_2 = 4.6817
$$
### Решение системы нелинейных уравнений 

Система:

$$  \begin{cases} 
    x^2 y^2 - 3 x^3 + 6 y^3 + 8 = 0 \\
    x^4 - 9y + 2 = 0
    \end{cases} $$

не имеет корней, поэтому была исследована система:

$$ \begin{cases}
    \sin{x} - y = 1.32 \\
    \cos{y} - x = -0.85
   \end{cases} $$

Локализация:

$$  x \in [\frac{\pi}{2}, 1.85] \\
    y \in [-\frac{\pi}{4}, 0]
$$

Критерием останова является условие, что модуль невязки меньше точности. 

Получен единственный корень:

$$ (x, y) = (1.79134, -0.34423) $$
